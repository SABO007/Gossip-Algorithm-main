```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
network_sizes = [10, 50, 100, 500, 1000, 5000]

gossip_full = [10, 30, 50, 150, 250, 500]
gossip_3d = [15, 40, 70, 200, 350, 700]
gossip_line = [20, 60, 100, 300, 500, 1000]
gossip_imp3d = [12, 35, 60, 180, 300, 600]

pushsum_full = [8, 25, 45, 130, 220, 450]
pushsum_3d = [12, 35, 65, 180, 320, 650]
pushsum_line = [18, 55, 90, 270, 450, 900]
pushsum_imp3d = [10, 30, 55, 160, 270, 550]
```

```python

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(network_sizes, gossip_full, 'o-', label='Full Network')
plt.plot(network_sizes, gossip_3d, 's-', label='3D Grid')
plt.plot(network_sizes, gossip_line, '^-', label='Line')
plt.plot(network_sizes, gossip_imp3d, 'D-', label='Imperfect 3D Grid')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Network Size')
plt.ylabel('Convergence Time (ms)')
plt.title('Gossip Algorithm')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(network_sizes, pushsum_full, 'o-', label='Full Network')
plt.plot(network_sizes, pushsum_3d, 's-', label='3D Grid')
plt.plot(network_sizes, pushsum_line, '^-', label='Line')
plt.plot(network_sizes, pushsum_imp3d, 'D-', label='Imperfect 3D Grid')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Network Size')
plt.ylabel('Convergence Time (ms)')
plt.title('Push-Sum Algorithm')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
