```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

gossip_full = {3000: 17378, 3500: 26285, 4000: 30900, 4500: 36525, 5000: 44011, 10000: 61436, 12000: 69784, 13000: 97037}
gossip_line = {1000: 2493, 2000: 2783, 3000: 3146, 4000: 3142, 5500: 5000, 6000: 5258, 5000: 3780}
gossip_3d = {3000: 8668, 5000: 11955, 6000: 11058, 8000: 10124, 10000: 14813, 12000: 14789, 14000: 15516, 16000: 14133, 25000: 20339, 50000: 49644, 75000: 69621, 100000: 110297, 200000: 1040167, 300000: 3403735}
gossip_imp3d = {50000: 33063, 100000: 70485, 150000: 311478, 200000: 445340}

pushsum_line = {100: 2033, 500: 222562, 600: 345746}
pushsum_full = {1000: 54, 5000: 1225, 10000: 6039, 15000: 38952}
pushsum_3d = {1000: 677, 10000: 43957, 25000: 225084, 50000: 683042}
pushsum_imp3d = {1000: 125, 5000: 894, 10000: 2573, 25000: 5993, 50000: 13035, 100000: 27929, 200000: 52271}
```

```python
def plot_data(ax, data_dict, label, marker):
    x = sorted(data_dict.keys())
    y = [data_dict[k] for k in x]
    ax.plot(x, y, label=label, marker=marker)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 8))

plot_data(ax1, gossip_full, 'Full Network', 'o')
plot_data(ax1, gossip_line, 'Line', 's')
plot_data(ax1, gossip_3d, '3D Grid', '^')
plot_data(ax1, gossip_imp3d, 'Imperfect 3D Grid', 'D')

ax1.set_xscale('log')
ax1.set_yscale('log')
ax1.set_xlabel('Number of Nodes')
ax1.set_ylabel('Convergence Time (ms)')
ax1.set_title('Gossip Algorithm: Convergence Time vs Network Size')
ax1.legend()
ax1.grid(True)

plot_data(ax2, pushsum_full, 'Full Network', 'o')
plot_data(ax2, pushsum_line, 'Line', 's')
plot_data(ax2, pushsum_3d, '3D Grid', '^')
plot_data(ax2, pushsum_imp3d, 'Imperfect 3D Grid', 'D')

ax2.set_xscale('log')
ax2.set_yscale('log')
ax2.set_xlabel('Number of Nodes')
ax2.set_ylabel('Convergence Time (ms)')
ax2.set_title('Push-Sum Algorithm: Convergence Time vs Network Size')
ax2.legend()
ax2.grid(True)

plt.tight_layout()
plt.show()
```
